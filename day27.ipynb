```python
# Import the required libraries
import requests
from bs4 import BeautifulSoup

# Fetch the web page
url = 'https://news.ycombinator.com/'  # Example URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    print("Successfully fetched the web page.")
    page_content = response.content
else:
    print("Failed to retrieve the web page. Status code:", response.status_code)

# Print a snippet of the page content for debugging
print(page_content[:2000])  # Print the first 2000 characters

# Parse the web page
soup = BeautifulSoup(page_content, 'html.parser')

# Extract and print titles of the news articles using the 'titleline' class
titles = soup.select('.titleline a')

if not titles:
    print("No titles found. The page structure might have changed.")
else:
    for i, title in enumerate(titles, start=1):
        print(f"{i}. {title.get_text()}")
```
